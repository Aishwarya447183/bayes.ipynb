```python
Q1.
Both t-test and z-test are statistical tests that help to determine the significance of differences between sample means and population means. However, they differ in their assumptions and the circumstances in which they can be used.

A z-test assumes that the population standard deviation is known, while a t-test assumes that the population standard deviation is unknown and uses an estimate of the standard deviation from the sample data. Therefore, a t-test is generally used when the sample size is small, and the population standard deviation is unknown.

A z-test is appropriate when the sample size is large (typically greater than 30) and the population standard deviation is known or can be assumed. It is used to test hypotheses about the mean of a normally distributed population when the population standard deviation is known.

For example, suppose a company is interested in comparing the salaries of two groups of employees: men and women. The company wants to know if there is a statistically significant difference in the mean salaries of these two groups. If the sample size is large (say, more than 30) and the population standard deviation is known, a z-test can be used to determine if there is a significant difference in the mean salaries.

On the other hand, if the sample size is small (say, less than 30) and the population standard deviation is unknown, a t-test is more appropriate. For instance, a researcher wants to test whether a new drug is effective in reducing blood pressure. The researcher takes a small sample of 20 participants and measures their blood pressure before and after taking the drug. A t-test can be used to determine if there is a significant difference in the mean blood pressure before and after taking the drug.
```

```python
Q2.
A statistical hypothesis test is a method of making decisions about a population based on sample data. In hypothesis testing, there are two types of tests: one-tailed tests and two-tailed tests.

In a two-tailed test, the null hypothesis is tested against an alternative hypothesis that suggests the parameter being tested is different from the null hypothesis value in either direction. This means that the rejection region is split between the two tails of the distribution. The critical values are determined by the significance level (usually α=0.05 or α=0.01) and the degrees of freedom.

For example, suppose we want to test the hypothesis that the mean height of students in a particular university is equal to 170 cm. The null hypothesis would be that the mean height of students is 170 cm, and the alternative hypothesis would be that the mean height is not equal to 170 cm. In this case, a two-tailed test would be appropriate because we are interested in whether the mean height is different from 170 cm in either direction (either taller or shorter).

In a one-tailed test, the null hypothesis is tested against an alternative hypothesis that suggests the parameter being tested is either greater than or less than the null hypothesis value. This means that the rejection region is located entirely in one tail of the distribution. The critical values are determined by the significance level, the degrees of freedom, and the direction of the test.

For example, suppose we want to test the hypothesis that a new drug reduces the blood pressure of patients. The null hypothesis would be that the drug does not reduce blood pressure, and the alternative hypothesis would be that the drug reduces blood pressure. In this case, a one-tailed test would be appropriate because we are only interested in whether the drug reduces blood pressure (i.e., in one direction, lower), not whether it has any effect in the opposite direction (higher).

In summary, a two-tailed test is used when the null hypothesis can be rejected in either direction, while a one-tailed test is used when the null hypothesis can only be rejected in one direction.
```

```python
Q3.
In hypothesis testing, there are two types of errors: Type 1 and Type 2 errors.

A Type 1 error occurs when the null hypothesis is rejected, even though it is actually true. In other words, a Type 1 error is a false positive result. The probability of making a Type 1 error is denoted by α and is usually set at 0.05 or 0.01. A smaller value of α reduces the risk of Type 1 errors, but it increases the risk of Type 2 errors.

For example, suppose a company wants to test the hypothesis that their new marketing strategy has increased sales by at least 10%. The null hypothesis would be that the new strategy has not increased sales by 10%, and the alternative hypothesis would be that it has. If the company rejects the null hypothesis and concludes that the new strategy has increased sales by 10%, but in reality, it has not, this would be a Type 1 error.

A Type 2 error occurs when the null hypothesis is not rejected, even though it is actually false. In other words, a Type 2 error is a false negative result. The probability of making a Type 2 error is denoted by β. The power of a test is equal to 1- β, and it measures the ability of the test to detect a true effect when it exists.

For example, suppose a medical researcher wants to test the hypothesis that a new drug reduces the incidence of a particular disease by at least 20%. The null hypothesis would be that the new drug does not reduce the incidence of the disease by 20%, and the alternative hypothesis would be that it does. If the researcher fails to reject the null hypothesis and concludes that the new drug has no effect, but in reality, it does reduce the incidence of the disease by 20%, this would be a Type 2 error.

In summary, a Type 1 error occurs when the null hypothesis is rejected when it is actually true, while a Type 2 error occurs when the null hypothesis is not rejected when it is actually false. The risk of making a Type 1 error can be controlled by setting the significance level, while the risk of making a Type 2 error can be reduced by increasing the sample size or by using a more powerful test.
```

```python
Q4.
Bayes's theorem is a mathematical formula that allows us to update the probability of an event occurring based on new evidence or information. The theorem is named after the Reverend Thomas Bayes, an 18th-century statistician and philosopher.

Bayes's theorem states that:

P(A|B) = P(B|A) * P(A) / P(B)

Where:
P(A|B) is the conditional probability of A given B
P(B|A) is the conditional probability of B given A
P(A) is the prior probability of A
P(B) is the prior probability of B

To understand how Bayes's theorem works, let's consider an example:

Suppose a certain disease affects 1 in 1000 people. There is a test for this disease that is 99% accurate: it correctly identifies 99% of people who have the disease and 99% of people who do not have the disease. If a person tests positive for the disease, what is the probability that they actually have the disease?

Let's define the following events:
A = the event that a person has the disease
B = the event that a person tests positive for the disease

We are interested in finding P(A|B), the probability that a person has the disease given that they test positive for it.

We can use Bayes's theorem to calculate this probability:

P(A|B) = P(B|A) * P(A) / P(B)

P(B|A) is the conditional probability that a person tests positive given that they have the disease. This is equal to 0.99, as the test correctly identifies 99% of people who have the disease.

P(A) is the prior probability that a person has the disease. This is equal to 0.001, as the disease affects 1 in 1000 people.

P(B) is the prior probability that a person tests positive for the disease. This is equal to the probability that a person has the disease and tests positive, plus the probability that a person does not have the disease and tests positive.

P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)
P(B) = 0.99 * 0.001 + 0.01 * 0.999
P(B) = 0.01098

Therefore, we can calculate the probability that a person has the disease given that they test positive for it as:

P(A|B) = P(B|A) * P(A) / P(B)
P(A|B) = 0.99 * 0.001 / 0.01098
P(A|B) = 0.0901

So the probability that a person has the disease given that they test positive for it is 0.0901 or approximately 9%. This means that even if a person tests positive for the disease, there is still a 91% chance that they do not have the disease.
```

```python
Q5.
A confidence interval is a range of values that is likely to contain the true value of a population parameter with a certain level of confidence. In statistical inference, we use a sample of data to estimate an unknown population parameter, such as the mean or proportion, and a confidence interval gives us an indication of the precision of that estimate.

The confidence level is the probability that the true population parameter falls within the confidence interval. It is typically expressed as a percentage, such as 95% or 99%.

To calculate a confidence interval, we need to know the sample size, the sample mean, and the standard error of the mean (SEM), which is a measure of the variability of the sample mean. The SEM can be calculated using the sample standard deviation and the sample size.

The formula for a confidence interval for the population mean (μ) is:

CI = X̄ ± z* (SEM)

Where:
X̄ is the sample mean
z is the z-score for the desired confidence level
SEM is the standard error of the mean

For example, let's say we want to calculate a 95% confidence interval for the average weight of a certain breed of dogs. We collect a random sample of 50 dogs from this breed and find that the sample mean weight is 25 pounds, and the sample standard deviation is 3 pounds.

First, we need to calculate the SEM:

SEM = s / sqrt(n)
SEM = 3 / sqrt(50)
SEM = 0.424

Next, we need to find the z-score for a 95% confidence level. We can look this up in a z-table or use a calculator, and we find that the z-score is 1.96.

Finally, we can calculate the confidence interval:

CI = 25 ± 1.96 * 0.424
CI = [24.16, 25.84]

Therefore, we can say with 95% confidence that the true population mean weight of this breed of dogs falls within the range of 24.16 to 25.84 pounds.
```
